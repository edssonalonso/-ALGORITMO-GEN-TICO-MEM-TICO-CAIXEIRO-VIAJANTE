#Problema d Caxeiro Viajante

Suponha que um caixeiro viajante tenha de visitar n cidades diferentes, iniciando e encerrando sua viagem na primeira cidade. Suponha, também, que não importa a ordem com que as cidades são visitadas e que de cada uma delas pode-se ir diretamente a qualquer outra.
O problema do caixeiro viajante consiste em descobrir a rota que torna mínima a viagem total.


Exemplificando o caso n = 4:

se tivermos quatro cidades A, B, C e D, uma rota que o caixeiro deve considerar poderia ser: saia de A e daí vá para B, dessa vá para C, e daí vá para D e então volte a A. Quais são as outras possibilidades ? É muito fácil ver que existem seis rotas possíveis:

- ABCDA
- ABDCA
- ACBDA
- ACDBA
- ADBCA
- ADCBA

###Order-Based Crossover (OBX)

image.png

```python
def Order_Based_Crossover(pai1, pai2, quant_genes):#dominio é o valor máximo assumido por cada elementos. quant_genes é a quant. de genes que vao sofrer mutacao
  posicoes_cross_over = (random.sample(range(len(pai1)),quant_genes))#Selecionando 20 posições/genes que vai sofrer mutacao
  posicoes_cross_over.sort()
  filho1 = []
  filho2 = []

  for i in range(len(pai1)):#criando os filhos sem cross_over
    if(posicoes_cross_over.count(i) == 1):# Ou seja, se ele entra neste IF, quer dizer que estamos num gene que vai sofrera mutacao
      filho1.append(-1)
      filho2.append(-1)
    else:
      filho1.append(pai1[i])
      filho2.append(pai2[i])                 # Add um cidade que nao existe nas posicoes que vao sofrer cross_over

  posicao_filho = 0
  for i in range(len(pai1)):
    if( filho1.count(pai2[i]) == 0 ):# se entramos nesse IF eh por que o PAI2 tem uma cidade que o filho1 n tem, logo add essa cidade no local de crossover
      filho1[posicoes_cross_over[posicao_filho]] = pai2[i]
      posicao_filho += 1

  posicao_filho = 0
  for i in range(len(pai1)):
    if( filho2.count(pai1[i]) == 0 ):# se entramos nesse IF eh por que o PAI2 tem uma cidade que o filho1 n tem, logo add essa cidade no local de crossover
      filho2[posicoes_cross_over[posicao_filho]] = pai1[i]
      posicao_filho += 1


  return filho1, filho2
```

###Position-Based Crossover (PBX)

image.png

```python
def Position_Based_Crossover(pai1, pai2, quant_genes):
  posicoes_cross_over = (random.sample(range(len(pai1)),quant_genes))#Selecionando 40% posições/genes que vai sofrer mutacao
  posicoes_cross_over.sort()
  filho1 = []
  filho2 = []

  for i in range(len(pai1)):#criando os filhos sem cross_over
    if(posicoes_cross_over.count(i) == 0):
      filho1.append(-1)
      filho2.append(-1)
    else:
      filho1.append(pai2[i])
      filho2.append(pai1[i])

  posicao_filho = 0
  for i in range(len(pai1)):

    while (filho1[posicao_filho] != -1):
      posicao_filho += 1
      if (posicao_filho >= len(filho1)):
        break

    if( filho1.count(pai1[i]) == 0 ):# se entramos nesse IF eh por que o PAI1 tem uma cidade que o filho1 n tem, logo add essa cidade no local de crossover
      if filho1[posicao_filho] == -1:
        filho1[posicao_filho] = pai1[i]
        posicao_filho += 1
    if (posicao_filho >= len(filho1)):
      break

  posicao_filho = 0
  for i in range(len(pai2)):

    while (filho2[posicao_filho] != -1):
      posicao_filho += 1
      if (posicao_filho >= len(filho2)):
        break

    if( filho2.count(pai2[i]) == 0 ):# se entramos nesse IF eh por que o PAI1 tem uma cidade que o filho1 n tem, logo add essa cidade no local de crossover
      if filho2[posicao_filho] == -1:
        filho2[posicao_filho] = pai2[i]
        posicao_filho += 1
    if (posicao_filho >= len(filho2)):
      break

  return filho1, filho2
```

###Partially Matched Crossover (PMX)

image.png

```python
def Partially_Matched_Crossover(pai1, pai2, quant_genes):
  inicio_intervalo = random.randint(0,len(pai1) - quant_genes)#elemento que dá inicio a o intervalo onde ocorrera o cross over
  interlavo_cross_over = [i for i in range(inicio_intervalo, quant_genes + inicio_intervalo)]
  filho1 = []
  filho2 = []
  mapeamento = []
  for i in range(len(pai1)):#criando os filhos sem cross_over e crinado mapeamento
    if(interlavo_cross_over.count(i) == 0):
      filho1.append(-1)
      filho2.append(-1)
    else:
      mapa = []
      mapa.append(pai2[i])
      mapa.append(pai1[i])
      filho1.append(pai2[i])
      filho2.append(pai1[i])
      mapeamento.append(mapa)

#CRIANDO MATRIZ ZERADA PARA QUARDAR AS DUPLAS DE MAPEAMENTO
  matriz_mapeamento = np.zeros((2,len(mapeamento) ))
  for i in range(len(mapeamento)):
    matriz_mapeamento[0][i] = mapeamento[i][0]
    matriz_mapeamento[1][i] = mapeamento[i][1]


  #AGRUPANDO MAPEAMENTO
  for i in range(len(matriz_mapeamento[0])):
    j = i
    mapa_parcial = np.array([])
    while(matriz_mapeamento[0][j] in matriz_mapeamento[1]):
      indice = int(np.where(matriz_mapeamento[1] == matriz_mapeamento[0][j])[0])
      if (indice == j) or (set(matriz_mapeamento[:,indice]).issubset(set(mapa_parcial))):
        break
      mapa_parcial = np.concatenate((mapa_parcial, matriz_mapeamento[:,j]), axis =None)
      mapa_parcial = np.concatenate((mapa_parcial, matriz_mapeamento[:,indice]), axis =None)
      j  = indice
      if j == len(matriz_mapeamento[0]):
        break

    mapeamento.append(np.unique(mapa_parcial).tolist())

  #Agrupamento começando por
  for i in range(len(matriz_mapeamento[0])):
    j = i
    mapa_parcial = np.array([])
    while(matriz_mapeamento[1][j] in matriz_mapeamento[0]):
      #print("np.where(matriz_mapeamento[0] == matriz_mapeamento[1][j])[0]: ", np.where(matriz_mapeamento[0] == matriz_mapeamento[1][j])[0])
      indice = int(np.where(matriz_mapeamento[0] == matriz_mapeamento[1][j])[0])
      if indice == j or (set(matriz_mapeamento[:,indice]).issubset(set(mapa_parcial)) ):
        break
      mapa_parcial = np.concatenate((mapa_parcial, matriz_mapeamento[:,j]), axis =None)
      mapa_parcial = np.concatenate((mapa_parcial, matriz_mapeamento[:,indice]), axis =None)
      j = indice
      if j == len(matriz_mapeamento[0]):
        break

    mapeamento.append(np.unique(mapa_parcial).tolist())

  def tamanho(lista):
    return len(lista)
  mapeamento.sort(key = tamanho)


  #verificar subgrupos dentro de grupos
  i = 0
  while(i < len(mapeamento)):
    flag = False
    j = i+1
    while j < len(mapeamento):
      if(set(mapeamento[i]).issubset(set(mapeamento[j]))):
        mapeamento.remove(mapeamento[i])
        flag = True
        break;
      j += 1
    if(flag):
      i = i;
    else:
      i += 1

  for i in range(len(pai1)):
    if(filho1[i] == -1):
      k = 0
      while(pai1[i] in filho1[inicio_intervalo: quant_genes + inicio_intervalo]):
        mapa = []
        for j in mapeamento:
          if(pai1[i] in j):
            mapa = j
            break
        pai1[i] = mapa[k]
        k += 1
      filho1[i] = pai1[i]

  for i in range(len(pai2)):
    if(filho2[i] == -1):
      k = 0
      while(pai2[i] in filho2[inicio_intervalo: quant_genes + inicio_intervalo]):
        mapa = []
        for j in mapeamento:
          if(pai2[i] in j):
            mapa = j
            break
        pai2[i] = mapa[k]
        k += 1
      filho2[i] = pai2[i]





  return filho1, filho2
```

```python
def Order_Based_Mutation(solucao):
  for i in range(int(len(solucao)*0.2)):
    genes = random.sample(range(len(solucao)),2)# vamos sortear duas POSICOES (genes/elementos) para sofrer a mutação
    mutante = solucao.copy()
    aux = mutante[genes[0]]
    mutante[genes[0]] = mutante[genes[1]]
    mutante[genes[1]] = aux
  return mutante
```

###Operador de Mutação (Position-Based Mutation)



image.png

```python
def Position_Based_Mutation(solucao):
  for i in range(int(len(solucao)*0.2)):
    genes = random.sample(range(len(solucao)),2)#Gerando 2 posicoes dentro do intervalo [0,49], a primeira posicao é o gene que vai ser deslocado
    genes.sort(reverse=False)
    mutante = solucao.copy()
    aux = mutante[genes[0]]
    for i in range(genes[0], genes[1]):
      if i == (genes[1] - 1):
        mutante[i] = aux
      else:
        mutante[i] = mutante[i + 1]
  return mutante
```

###Operador de Mutação (Scramble Mutation)

image.png

```python
def Scramble_Mutation(solucao):
  inicio_intervalo = random.randint(0,len(solucao) - int(len(solucao)*0.4))#elemento que dá inicio a o intervalo onde ocorrera o cross over
  interlavo_mutacao = [i for i in range(inicio_intervalo, int(len(solucao)*0.4) + inicio_intervalo)]
  solucao_aux = solucao.copy()
  aux = interlavo_mutacao.copy()
  random.shuffle(aux)
  j = 0
  for i in interlavo_mutacao:
    solucao[i] = solucao_aux[aux[j]]
    j += 1
  return solucao
```

#Função Custo

```python
def funcao_custo_minimizacao(solucao, cidades):#Vai receber uma solucao e vi devolver a distância percorrida
  distancia_percorrida = 0
  #print("solucao: ", solucao)
  #print("len(cidades): ", len(cidades))
  for i in range(len(solucao)):
    #print("i: ",i, " solucao[i]: ", solucao[i], " len(solucao): ", len(solucao))
    if i == (len(solucao)-1):
      dist_x = np.absolute(cidades[int(solucao[i])][0] - cidades[int(solucao[0])][0])
      dist_y = np.absolute(cidades[int(solucao[i])][1] - cidades[int(solucao[0])][1])
      distancia_percorrida += np.sqrt(dist_x**2 + dist_y**2)
    else:
      dist_x = np.absolute(cidades[int(solucao[i])][0] - cidades[int(solucao[i+1])][0])
      dist_y = np.absolute(cidades[int(solucao[i])][1] - cidades[int(solucao[i+1])][1])
      distancia_percorrida += np.sqrt(dist_x**2 + dist_y**2)
  return (9999999999 - distancia_percorrida)
```

[ (solucao01, custo01), (solucao02, custo02), ..., (solucao50, custo50)]

Lembrando que uma solucao é uma lista de 50 posicoes onde cada posicao tem dentro um valor entre 0 á 49 (que representa as cidades), a o custo é o custo dessa solucao, ou seja, é a distância dessa solucao.

```python
def genetico_modificado(funcao_custo_minimizacao, tamanho_populacao,
             prob_mutacao, numero_geracoes, numero_elitismo, tamanho_solucao, cidades):


  def custo (tupla):
    return tupla[0]

  historico_melhores_solucoes = []
  historico_media_geracao = []
  numero_geracoes_atingido = 0
  limitador_geracoes = 0

  populacao_virgem = []
  for i in range(tamanho_populacao):#cria populacaao inicial randomica
    solucao = random.sample(range(tamanho_solucao),tamanho_solucao)
    populacao_virgem.append(solucao)

  custos_populacao = [(funcao_custo_minimizacao(individuo, cidades), individuo) for individuo in populacao_virgem]#avalia cada individuo (calcula fitness)
  custos_populacao.sort(key = custo, reverse = True)
  populacao = custos_populacao

  for i in range(numero_geracoes):#avaliando o custo de cada solucao/individuo da populacao gerada
    numero_geracoes_atingido = i
    populacao = seleciona_pais(populacao, numero_elitismo, int(tamanho_populacao/2))#ASeleciona os 10 melhores pais e mais 10 com o algoritmo selecao por janela.
    historico_melhores_solucoes.append(populacao[0])

    custos_coluna = [solucao[0] for solucao in populacao]
    historico_media_geracao.append(np.mean(custos_coluna))

    while len(populacao) < tamanho_populacao:#a populacao tem 20 individuos. Vamos gerar filhos (ou seja a pop. vai crecer) até que ela chegue em 50 individuos. Não matamos os pais
      if random.random() <= prob_mutacao:
        randomico = random.random()#valor randomico entre 0 e 1
        m = random.sample(populacao, 1)#selecionando uma solucao dentro da populcao, ele vai sofrer a mutacao

        if(randomico <= 0.3):
          mutante = Scramble_Mutation(m[0][1].copy())
          populacao.append((funcao_custo_minimizacao(mutante, cidades), mutante))#selecionando aleatoriamente 20 genes para sofrer mutacao
        elif(randomico > 0.3 and randomico <= 0.6):
          mutante = Position_Based_Mutation(m[0][1].copy())
          populacao.append((funcao_custo_minimizacao(mutante, cidades),mutante))
        else:
          mutante = Order_Based_Mutation(m[0][1].copy())
          populacao.append((funcao_custo_minimizacao(mutante, cidades),mutante))

      else: #Realizando Cross Over entre dois pais
        randomico = random.random()#valor randomico entre 0 e 1
        pai1 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over
        pai2 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over
        while(pai2[0][1] == pai1[0][1]):#Obrigando garantindo que os pais serao diferentes
          pai2 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over

        if(randomico <= 0.3):
          filho1, filho2 = Partially_Matched_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))
        elif(randomico > 0.3 and randomico <= 0.6):
          filho1, filho2 = Position_Based_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))
        else:
          filho1, filho2 = Order_Based_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))

    populacao.sort(key = custo, reverse= True)
    if((i > 0) and (historico_melhores_solucoes[i][0] == historico_melhores_solucoes[i-1][0])):
      limitador_geracoes += 1
    else:
      limitador_geracoes = 0
    if limitador_geracoes > 50:
      prob_mutacao = 0.2
    if limitador_geracoes > 100:
      prob_mutacao = 0.3
    if limitador_geracoes > 150:
      break


  return populacao[0], historico_melhores_solucoes, historico_media_geracao, numero_geracoes_atingido
```

```python
def distancia(x1, x2, y1, y2):#Vai receber uma solucao e vi devolver a distância percorrida
  distancia = 0
  dist_x = np.absolute(x1 - x2)
  dist_y = np.absolute(y1 - y2)
  distancia = np.sqrt(dist_x**2 + dist_y**2)

  return distancia
```

```python
def Operador_Vizinho_Mais_Proximo(solucao, cidades):
    nova_solucao = solucao.copy()

    posicao_cidade = random.randint(0,len(solucao) - 1)
    valor_cidade = nova_solucao[posicao_cidade]
    nova_solucao.pop(posicao_cidade)

    # achar vizinho mais próximo
    x = cidades[int(valor_cidade)][0]
    y = cidades[int(valor_cidade)][1]

    vizinho = -1
    distancia_vizinho = 999999999999

    for i in range(len(solucao)):
        #print("solucao[i]: ", solucao[i])
        tmp_x = cidades[int(solucao[i])][0]
        tmp_y = cidades[int(solucao[i])][1]

        tmp_distancia = distancia(x, tmp_x, y, tmp_y)
        if distancia_vizinho > tmp_distancia:
            vizinho = i
            distancia_vizinho = tmp_distancia

    nova_solucao.insert(vizinho+1, valor_cidade)

    # verificar melhor solucação
    custo_solucao = funcao_custo_minimizacao(solucao, cidades)
    custo_nova_solucao = funcao_custo_minimizacao(nova_solucao, cidades)

    #print(solucao)
    #print(nova_solucao)
    #if custo_solucao > custo_nova_solucao:
        #return custo_solucao
    #else:
    return nova_solucao
```

```python
def genetico_modificado_memetico(funcao_custo_minimizacao, tamanho_populacao,
             prob_mutacao, numero_geracoes, numero_elitismo, tamanho_solucao, cidades):

  def custo (tupla):
    return tupla[0]

  historico_melhores_solucoes = []
  historico_media_geracao = []
  limitador_geracoes = 0

  populacao_virgem = []
  for i in range(tamanho_populacao):#cria populacaao inicial randomica
    solucao = random.sample(range(tamanho_solucao),tamanho_solucao)
    populacao_virgem.append(solucao)

  custos_populacao = [(funcao_custo_minimizacao(individuo, cidades), individuo) for individuo in populacao_virgem]#avalia cada individuo (calcula fitness)
  custos_populacao.sort(key = custo, reverse = True)
  populacao = Otimizacao02(custos_populacao, cidades)


  for i in range(numero_geracoes):#avaliando o custo de cada solucao/individuo da populacao gerada

    if i == 0:
      populacao = seleciona_pais(populacao, 20, 50)
    else:
      populacao = seleciona_pais(populacao, numero_elitismo, 100)#ASeleciona os 10 melhores pais e mais 10 com o algoritmo selecao por janela.

    historico_melhores_solucoes.append(populacao[0])

    custos_coluna = [solucao[0] for solucao in populacao]
    historico_media_geracao.append(np.mean(custos_coluna))

    while len(populacao) < tamanho_populacao:#a populacao tem 20 individuos. Vamos gerar filhos (ou seja a pop. vai crecer) até que ela chegue em 50 individuos. Não matamos os pais

      if random.random() <= prob_mutacao:
        randomico = random.random()#valor randomico entre 0 e 1
        m = random.sample(populacao, 1)#selecionando uma solucao dentro da populcao, ele vai sofrer a mutacao

        if(randomico <= 0.3):
          mutante = Scramble_Mutation(m[0][1], int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(mutante, cidades), mutante))#selecionando aleatoriamente 20 genes para sofrer mutacao
        elif(randomico > 0.3 and randomico <= 0.6):
          mutante = Position_Based_Mutation(m[0][1])
          populacao.append((funcao_custo_minimizacao(mutante, cidades),mutante))
        else:
          mutante = Order_Based_Mutation(m[0][1])
          populacao.append((funcao_custo_minimizacao(mutante, cidades),mutante))

      else: #Realizando Cross Over entre dois pais

        randomico = random.random()#valor randomico entre 0 e 1
        pai1 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over
        pai2 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over
        while(pai2[0][1] == pai1[0][1]):#Obrigando garantindo que os pais serao diferentes
          pai2 = random.sample(populacao, 1)#Selecionando pai que vai fazer cross over

        if(randomico <= 0.3):
          filho1, filho2 = Partially_Matched_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))
        elif(randomico > 0.3 and randomico <= 0.6):
          filho1, filho2 = Position_Based_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))
        else:
          filho1, filho2 = Order_Based_Crossover(pai1[0][1].copy(), pai2[0][1].copy(), int(tamanho_solucao*0.4))
          populacao.append((funcao_custo_minimizacao(filho1, cidades), filho1))
          populacao.append((funcao_custo_minimizacao(filho2, cidades), filho2))


    populacao = Otimizacao02(populacao, cidades)
    populacao.sort(key = custo, reverse= True)


    if((i > 0) and (historico_melhores_solucoes[i][0] == historico_melhores_solucoes[i-1][0])):
      limitador_geracoes += 1
    else:
      limitador_geracoes = 0
    if limitador_geracoes > 50 and limitador_geracoes <= 100:
      prob_mutacao = prob_mutacao + 0.1
    elif limitador_geracoes > 100 and  limitador_geracoes < 150 :
      prob_mutacao = prob_mutacao + 0.2
    if limitador_geracoes >= 150:
      break

  return populacao[0][0], historico_melhores_solucoes, historico_media_geracao
```
